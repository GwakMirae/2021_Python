```python
def input_grade():
    arr = []
    for i in range(0,5):
        a = input('학점: ')
        arr.append(a)
    print('학점 입력 완료!')
    return arr
m = input_grade()
r = input_grade()
print(m, r)
```

```python
def total_grade(arr):
    total = sum(map(float,arr))
    grade = total/len(arr)
    return grade
m_g = total_grade(m)
r_g = total_grade(r)
print(m_g, r_g)
```

```python
import matplotlib.pyplot as plt
newM = list(map(float,m))
newR = list(map(float,r))
plt.title('grade')
plt.plot([1,2,3,4,5],newM)
plt.plot([1,2,3,4,5],newR)
plt.xlabel('subject')
plt.ylabel('grade')
plt.axis([0,6,0,5])
plt.legend(['m','r'])
plt.show()
```
